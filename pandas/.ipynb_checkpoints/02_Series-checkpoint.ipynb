# Series

```python
# 경고 메시지 무시
import warnings
warnings.filterwarnings(action='ignore') 

import pandas as pd
import numpy as np
```

### 1. Series 1차원 배열
- 판다스의 1차원 배열 구조를 가짐

```python
s1 = pd.Series([1, 2, 3])
s1
```

```python
array1 = np.array([1, 2, 3, 4, 5])
s1 = pd.Series(array1)
s1
```

```python
array1 = np.array([1, 2, 3, 4, 5])
s1 = pd.Series(array1 , index = ['a','b', 'c', 'd', 'e'])
s1
```

```python
dict1 = {"number" : [1, 2, 3]}
s1 = pd.Series(dict1)
s1
```

```python
dict2 = {
        'apple': '🍎',
        'orange' : '🍊',
        'lemon' : '🍋'
        }
s1 = pd.Series(dict2)
s1.index
```

```python
s = pd.Series([1, 2, 3])
s
```

```python
# 사이즈 확인 - 시리즈.size
print(len(s), s.size)
```

```python
# 사이즈 형태 확인
s.shape
```

```python
# 사이즈의 데이터 타입 확인
s.dtype
```

```python
# 사이즈의 인덱스 확인
print(f"인덱스 범위 : {s.index}")
print(f"인덱스 값 : {s.index.values}")
```
