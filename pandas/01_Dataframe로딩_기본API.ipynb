# 2. Dataframe

```python
## 파이썬의 2차원 리스트 생성
list1 = [['김철수', 28, '개발자'], ['이영희', 34, '디자이너']]
```

```python
# 리스트를 데이터프레임으로 생성
list_df = pd.DataFrame(list1, columns = ['이름', '나이', '직업'])
list_df
```

```python
# 딕셔너리로 데이터프레임 생성
dict_df = pd.DataFrame(dic1)
dict_df
```

# 4. ✨ info()
- 데이터프레임의 컬럼명, 데이터 타입, NaN(Null)건수, 데이터 건수의 정보를 출력

```python
# 1. 나이는 평균값으로 채우기
titanic_df['Age'] = titanic_df['Age'].fillna(titanic_df['Age'].mean())

# 2. 선실은 'Unknown'으로 채우기
titanic_df['Cabin'] = titanic_df['Cabin'].fillna('Unknown')

# 3. 탑승 항구는 몇 개 안 되니 NaN이 있는 행 삭제하기
titanic_df = titanic_df.dropna(subset=['Embarked'])

# 결과 확인 (모든 컬럼의 개수가 동일해집니다)
titanic_df.info()
```

# 6. value_counts()
- 동일한 개별 데이터 값이 몇건이 있는지 정보를 제공한다.

```python
tt_value_counts = titanic_df['Pclass'].value_counts()
print(tt_value_counts, type(tt_value_counts))
```

```python
tt_value_counts = titanic_df['Survived'].value_counts()
print(tt_value_counts, type(tt_value_counts))
```
