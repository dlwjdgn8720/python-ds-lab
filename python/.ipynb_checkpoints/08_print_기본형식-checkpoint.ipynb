# print()

## 1. 기본 문법
- print(value, ..., sep='구분기호', end= '\n', flush=False)

```python
a = 5
a
```

```python
b = 3
b
```

```python
a + b
```

```python
print(a)
print(b)
print(a+b)
```

```python
a + b
```

```python
print(1, 2, 3, sep='-') #1-2-3
```

```python
print(1, 2, 3, sep='-', flush=True) # 1-2-3
```

```python
print(a, end='\t')
print(b, end='\t')
print(a+b)
```

```python
name = '홍길동'
age = 30
#1. 콤마(,)를 사용하여 문자열 연결
print('이름:', name, '나이:', age)
```

```python
#2. 플러스 기호(+)를 사용하여 문자열 연결, 문자열만 가능
print('이름:'+ name, '나이:'+ str(age)) # int() - 숫자로 변환, str()- 문자열로 변환
```

```python
#3. % 포맷(old version)
print("이름: %s", "")
```
