## file 열기(open())
- 내장 함수인 open() 함수를 이용하여 파일을 연 다음, Read, Write가 가능함.
- 형식 : file = open(파일주소)
- 속성
   - r : 읽기(read) 모드로 파일 열기 (기본)
   - w : 쓰기(write) 모드로 파일 열기
   - a : 추가(append) 모드로 파일 열기, 같은 이름의 파일이 없으면 w 모드로 생성됨
   - b : 바이너리 파일 모드로 열기
   - t : 텍스트 파일 모드로 열기

### 1. 파일 쓰기
- 파일 쓰기를 하려면, open() 함수의 쓰기 모드로 열어야 함
- 파일 쓰기가 종료하면, 파일을 닫아야함

```python
# myFile.txt 파일을 생성
f = open('myFile.txt', 'w')
f.write('Hello~ myFile~ ^^')
f.close()
```

### 2. 파일 읽기
- 파일 읽기를 하려면, open() 함수의 읽기(r) 모드로 열어야 함
- 파일 읽기가 종료하면, 파일을 닫아야함

```python
# myFile.txt 파일 읽기
f = open('myFile.txt', 'r')
f_text = f.read()
f.close()

f_text
```

```python
# myFile_ko.txt 파일 생성(인코딩)
f = open('myFile_ko.txt', 'w', encoding='utf-8')
f.write('안녕~ 파이썬 ^^')
f.close()
```

```python
# myFile_ko.txt 파일 읽기
f = open('myFile_ko.txt', 'r')
# print(f.read())
text = f.read()
f.close()

text
```

```python
import os 
if os.path.exists('myFile_ko.txt'):
    os.remove('myFile_ko.txt')
    print("파일이 성공적으로 삭제되었습니다!")
else:
    print("지울 파일이 없습니다.")
```

## 3. 파일에 문자열 한 줄씩 쓰기

```python
# 파일에 문자열 한 줄씩 쓰기
f = open('myFile_multi.txt', 'w')
for num in range(1, 6):
    str = f"num: {num} \n"
    f.write(str)
f.close()

!type myFile_multi.txt
```

```python
# myFile_multi.txt 읽기, readLine() : 한 줄씩 읽기
f = open('myfile_multi.txt', 'r')
line1 = f.readline()
line2 = f.readline()
f.close()
print(line1)
print(line2)
```

```python
# myFile_multi.txt 파일을 반복적으로 읽기
f = open('myFile_multi.txt')
line = f.readline() # 맨 위의 데이터 읽어오기
while line:
    print(line)
    line = f.readline()
print("-" * 10)
f.close()
```
