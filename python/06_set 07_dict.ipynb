# 세트(set)
- 수학의 집합개념을 구현한 데이터타입
- 데이터에 순서가 없고, 중복해서 사용이 불가
- 교집합, 합집합, 차집합을 구하는 메서드

## 1. 세트 생성

```python
set1 = {1, 2, 3}
```

```python
type(set1)
```

```python
fruits_set = {'Apple', 'Lemon', 'Apple', 'Lemon'} #데이터 중복 불가
fruits_set
```

## 2. 세트(집합) 연산

* 교집합
  - set객체.intersection(set객체)
  - &

* 합집합
    - set객체.union(set객체)
    - |
* 차집합
    - set겍체.difference(set객체)

```python
# 세트 정의
set1 = {1, 2, 3}
set2 = {2, 3, 4}
```

```python
# 교집합
print(set1.intersection(set2), set1&set2) # 교집합 세트를 생성하여 리턴
```

```python
# 합집합
print(set1.union(set2), set1|set2)
```

```python
#차집합
print(set1.difference(set2), set2.difference(set1))
```

## 6. 리스트, 튜플, 세트 간 변환 타입
- list(), tuple(),set()

```python
a = [1, 2, 3, 4, 5]
b = (1, 2, 3, 4, 5)
c = {1, 3, 3, 4, 5}
```

```python
print(type(a), type(b), type(c))
```

```python
# a => tuple, set
a_to_tuple = tuple(a)
a_to_tuple
```

```python
# b => list, set
b_to_set = set(b)
b_to_set.remove(4)
```

```python
b_to_set
```

```python
# c => list,tuple

c_to_list = list(c)
c_to_list.append(5)
c_to_list
```

```python
c_to_set = set(c_to_list)
c_to_set
```

## 딕셔너리
- 키(key), 값(value)의 쌍으로 구성된 데이터 타입

- 키(key) : 임의로 지정한 숫자나 문자열

- 값(value): 어떤 데이터 타입도 가능

- 중괄호 {키:값, 키:값, ... }

## 1. 딕셔너리 생성

```python
dict1 = {
    'name': '홍길동',
    'age': 30,
    'address' : '서울시 강남구'
}
```

```python
print(dict1)
```

```python
fruits = {
    "name": "apple",
    "color": "red",
    "emoji": "🍎"
}
```

```python
fruits
```

## 2. 아이템 선택/추가
- 딕셔너리 key를 이용한 데이터 변경
- 추가/수정 : update()

```python
# 딕셔너리 객체의 모든 키 리턴
dict1_keys = dict1.keys()
type(dict1_keys)
dict1_keys
```

```python
fruits.keys()
```

```python
fruits['emoji']
```

```python
fruits['emoji']  = '🍏'
fruits
```

```python
# 딕셔너리 update함수를 이용한 데이터 변경 - {key : value}
fruits.update({"color" : "green"})
fruits
```

- copy(): 딕셔너리 객체 복사 : JS의 deep copy 형식으로 독립적인 객체로 생성

```python
dict1 = {'korea': '서울', 'japan':'도쿄'}
```

```python
dict1
```

```python
dict2 = dict1.copy() # 메모리 영역이 독립적인 객체로 생성, 원본(dict1)에 영향을 주지 않음
```

```python
print(dict1, dict2)
```

```python
# dict2의 korea 아이템 삭제
dict2.pop('korea')
```

```python
print(dict1, dict2)
```

```python
dict3 = dict1 # JS의 shallow copy 형식으로 주소값을 공유함, 원본에 영향을 줌
```

```python
print(dict1, dict3)
```

```python
dict3.pop('korea')
```

```python
print(dict1, dict3)
```
