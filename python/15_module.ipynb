# Module(모듈)
- 코드와 함수가 지정된 .py 파일
- 파이썬(.py) 파일로 생성된 모듈은 import 통해 사용함
- 장점
  - 생성된 코드의 재사용성을 높임
  - 공동작업시 유리함
  - 관리가 수월함
- 모듈 생성
  - 코드 이름 '모듈이름.py'로 저장
  - 주피터 노트북 사용시 내장함수 '%%writefile' 코드를 사용함, 반드시 맨 처음 작성

```python
%%writefile calculator.py

def add(a, b): return a + b;
def sub(a, b): return a - b;
def mul(a, b): return a * b;
def div(a, b): return a / b;

def gugudan(dan):
    for k in range(1,10):
        print(f"{dan} ✖ {k} = {dan*k}")
```

```python
# 모듈 import
import calculator as cal
import importlib
importlib.reload(cal)

a = 10
b = 5
print(cal.add(a, b))
print(cal.sub(a, b))
print(cal.mul(a, b))
print(cal.div(a, b))
cal.gugudan(3)
```
